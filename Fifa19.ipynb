```r
install.packages("FactoMineR")
```

```r
fifa <- read.csv(file = 'Fifa 2019.csv', row.names = 2)
```

```r
summary(fifa)
```

```r
num_cols <- unlist(lapply(fifa, is.numeric))
fifa_num <- fifa[, num_cols]
colnames(fifa_num)
```

```r
library(FactoMineR)
fifa_pca <- PCA(fifa_num,
                scale.unit = TRUE,
               ncp = 5)
```

```r
plot(fifa_pca, axes=c(1,2))
```
